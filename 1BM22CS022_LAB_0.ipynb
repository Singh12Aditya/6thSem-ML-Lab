```python
# Step 1: Import required libraries
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tickers = ["RELIANCE.NS", "TCS.NS", "INFY.NS"]
```

```python
# Fetch historical data for the last 1 year
data = yf.download(tickers, start="2022-10-01", end="2023-10-01", group_by='ticker')
```

```python
# Display the first 5 rows of the dataset
print("First 5 rows of the dataset:")
print(data.head())
```

```python
# Check the shape of the dataset
print("\nShape of the dataset:")
print(data.shape)

# Check column names
print("\nColumn names:")
print(data.columns)

# Summary statistics for a specific stock (e.g., Reliance)
reliance_data = data['RELIANCE.NS']
print("\nSummary statistics for Reliance Industries:")
print(reliance_data.describe())

# Calculate daily returns
reliance_data['Daily Return'] = reliance_data['Close'].pct_change()
```

```python
# Calculate daily returns
reliance_data['Daily Return'] = reliance_data['Close'].pct_change()
```

```python
# Plot the closing price and daily returns
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
reliance_data['Close'].plot(title="Reliance Industries - Closing Price")
plt.subplot(2, 1, 2)
reliance_data['Daily Return'].plot(title="Reliance Industries - Daily Returns", color='orange')
plt.tight_layout()
plt.show()
```

```python
# Save the Reliance data to a CSV file
reliance_data.to_csv('reliance_stock_data.csv')
print("\nReliance stock data saved to 'reliance_stock_data.csv'.")
```

```python
# Step 1: Import required libraries
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Define stock tickers and date range
tickers = ["HDFCBANK.NS", "ICICIBANK.NS", "KOTAKBANK.NS"]
start_date = "2024-01-01"
end_date = "2024-12-30"

# Step 3: Download stock data
stock_data = yf.download(tickers, start=start_date, end=end_date)

# Step 4: Extract closing prices
closing_prices = stock_data["Close"]

# Step 5: Plot closing prices
plt.figure(figsize=(12, 6))
for ticker in tickers:
    plt.plot(closing_prices.index, closing_prices[ticker], label=ticker)

plt.title("Closing Prices of HDFC Bank, ICICI Bank, and Kotak Mahindra Bank")
plt.xlabel("Date")
plt.ylabel("Closing Price (INR)")
plt.legend()
plt.grid()
plt.show()

# Step 6: Calculate daily returns
daily_returns = closing_prices.pct_change()

# Step 7: Plot daily returns
plt.figure(figsize=(12, 6))
for ticker in tickers:
    plt.plot(daily_returns.index, daily_returns[ticker], label=ticker)

plt.title("Daily Returns of HDFC Bank, ICICI Bank, and Kotak Mahindra Bank")
plt.xlabel("Date")
plt.ylabel("Daily Return")
plt.legend()
plt.grid()
plt.show()
```
